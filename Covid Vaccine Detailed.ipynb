```python
sns.set_style('darkgrid')
plt.figure(figsize=(90,10))

sns.countplot("vaccines", data=vaccine, order=vaccine["vaccines"].value_counts().index)

plt.tight_layout()
plt.ylabel("# of Countries", fontsize=12)
plt.xlabel("Vaccine/Combination of Vaccines", fontsize=12)
```

```python
county_vac2 = df_cleaned.groupby(["country",'iso_code','vaccines'])['total_vaccinations','people_vaccinated','people_fully_vaccinated',
                                           'daily_vaccinations','total_vaccinations_per_hundred','people_vaccinated_per_hundred',
                                           "people_fully_vaccinated_per_hundred",'daily_vaccinations_per_million'].max().reset_index()
county_vac2.head(10)
```

```python
#Visualisation of Country VS Total Vaccination

plt.figure(figsize=(20, 6))
county_vac2.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)[:50].plot(kind='bar');
plt.title('Country vs Total Vaccination', fontsize=24, fontweight='bold')
plt.ylabel('Total Vaccinations');
```

```python
#Visualisation of Country VS People Vaccinate

plt.figure(figsize=(20, 6))
data = county_vac2.groupby('country')['people_vaccinated'].max().sort_values(ascending=False)
plt.bar(data.index, data);
plt.xticks(rotation='90')
plt.yticks()
plt.title('Country vs People Vaccinated', fontsize=24, fontweight='bold')
plt.ylabel('people vaccinated');
```

```python
#Visualisation for daily vaccination per million in countries

plt.figure(figsize=(20, 6))
county_vac2.groupby('country')['daily_vaccinations_per_million'].max().sort_values(ascending=False)[:50].plot(kind='bar')
plt.ylabel('daily vaccinations per million')
plt.title('Daily Vaccination in Country per million', fontsize=24, fontweight='bold');
```

```python
#Visualisation for no. of people fully vaccinated in countries

plt.figure(figsize=(20, 6))
county_vac2.groupby('country')['people_fully_vaccinated'].max().sort_values(ascending=False)[:50].plot(kind='bar')
plt.ylabel('People fully vaccinated')
plt.title('People fully vaccinated in Country', fontsize=24, fontweight='bold');
```

```python
#Visualisation for Top 10 countries vaccinations (sorted by mean values)

df3 = df_cleaned.loc[:,['country', 'date', 'vaccines', 'daily_vaccinations']].dropna(subset=['daily_vaccinations'])
df3["date"] = pd.to_datetime(df3["date"], format = '%Y-%m-%d')

df3_agg = df3.groupby('country').daily_vaccinations.agg(
    sum_vaccination  = 'sum',
    mean_vaccination = 'mean')


final_df1 = df3.merge(df3_agg, on = 'country')
final_df1
```

```python
#which country has vaccinated more people

total_vaccinations = county_vac.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)
total_vaccinations1 = pd.DataFrame(total_vaccinations).reset_index()
total_vaccinations2 = total_vaccinations1.iloc[:15, :]
total_vaccinations2
```

```python
size_label=12
size_ticks = 10
size_title= 20

plt.figure(figsize=(12,10))
sns.barplot('total_vaccinations','country', data=total_vaccinations2, orient='h', palette="rocket")

plt.title('Top 15 countries that have given most vaccinations on {13th April 2021}', size=size_title)
plt.xlabel('Country', size=size_label)
plt.ylabel('Number of vaccinations', size=size_label)
plt.xticks(size=size_ticks)
plt.yticks(size=size_ticks)

plt.show()
```

```python
#Average daily vaccinations by country - Top 10

df_cleaned.groupby(['country']).agg({'daily_vaccinations':'mean'}).sort_values(by='daily_vaccinations', ascending=False)[:10].plot.barh(color="blue", figsize=(15,5))
plt.title('Average daily vaccinations by country - Top 10')
plt.legend('')
plt.xlabel('average daily vaccinations')
plt.show()
```

```python
sort = df3_agg.sort_values(by = 'mean_vaccination', ascending = False).reset_index()
countries = sort.country.to_list()

graphing_mean = final_df1.sort_values(by = 'mean_vaccination', ascending = False)
top10_mean = graphing_mean[graphing_mean['country'].isin(countries[:10])]

fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot()
sns.lineplot( x = 'date', y = 'daily_vaccinations', hue = 'country', 
    data = top10_mean)
plt.legend(ncol = 3, frameon = False, title = '')
plt.xticks(rotation=45) 
plt.title('Top 10 countries vaccinations (sorted by mean values)')
```

```python
#Timeline of vaccinations for selected countrie

selected_countries = ['India', 'Germany', 'United States', 'United Arab Emirates']
data_countries = df_cleaned[df_cleaned['country'].isin(selected_countries)].sort_values(by='date', ascending=True)


plt.figure(1, figsize=(20,10))
sns.lineplot(x='date',y='daily_vaccinations', hue='country', data = data_countries)
plt.xticks(rotation=70)
plt.title('Timeline of vaccinations for selected countries', size=20)
plt.legend()
plt.show()
```
